```python
cast_df = pd.DataFrame()
cast_df = df['cast'].str.split(',',expand = True).stack()
cast_df = cast_df.to_frame()
cast_df.columns = ['Actors']
actors = cast_df.groupby(['Actors']).size().reset_index(name = "total_no")
actors = actors[actors.Actors != 'actor not specified']
acotrs = actors.sort_values(by = "total_no")
top5_actors = actors.head()
top5_actors = top5_actors.sort_values(by = 'total_no')
bar_actors = px.bar(top5_actors , x = 'total_no' , y = 'Actors',title = 'Netflix top5 actors')
bar_actors
```

```python
df1 = df[['release_year','type']]
df1 = df1.rename(columns = {'release_year':'Release_year','type':'Type'})
df2 = df1.groupby(['Release_year','Type']).size().reset_index(name = "total_count")
df2.head()
```

# sentiment analysis of the content


```python
df3 = df[['release_year','description']]
df3 = df3.rename(columns = {'release_year':'Release_year','description':'Description'})
for index , row in df3.iterrows():
  d = row['Description']
  testimonial= TextBlob(d)
  p = testimonial.sentiment.polarity
  if p==0:
    sent = 'neutral'
  elif p > 0 :
    sent = 'positive'
  else :
    sent = 'negative'
  df3.loc[[index , 2],'sentiment'] = sent
df3 = df3.groupby(['Release_year','sentiment']).size().reset_index(name = "Total_count")
df3 = df3[df3['Release_year']>2005]
graph = px.bar(df3 , x = "Release_year",y = 'Total_count',color = 'sentiment')
graph.show()




```
